STEP 2: perform diarization

```python
from pyannote.audio import Pipeline
import torch

HUGGINGFACE_TOKEN = "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"

pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization-3.1",
    use_auth_token=HUGGINGFACE_TOKEN
)

pipeline.to(torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu"))
```

STEP 3: store the diarization result in a csv file in the desired directory

```python
cd /home/zmi/data/train_L/wav
```

```python
import os

directory_path = "/data/zmi/train_L/wav"

audio_file_paths = [os.path.join(directory_path, file) for file in os.listdir(directory_path)]
```

```python
results = []
for audio in audio_file_paths:
    print(f"Processing file: {audio}")
    diarization = pipeline(audio)
    file_name = os.path.basename(audio
                                )
    for turn, _, speaker in diarization.itertracks(yield_label=True):
        results.append({
            "file_name": file_name,
            "start_time": turn.start,
            "end_time": turn.end,
            "speaker": speaker
        })
        
```

```python
import pandas as pd
diarization_result_df = pd.DataFrame(results)

diarization_result_df.to_csv("diarization_results.csv", index=False)
```
